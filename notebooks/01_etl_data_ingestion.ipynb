```python
# Notebook: 01_etl_data_ingestion.py

# 1. Definindo o Caminho de Origem (Acessível pelo Spark)
data_path = "dbfs:/Volumes/dev_catalogue/staging_schema/raw_data_volume/insurance.csv"

# 2. Leitura Direta pelo Spark (Totalmente Distribuída e Segura)
df_raw_spark = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(data_path)
)

print("Schema do DataFrame Bruto:")
df_raw_spark.printSchema()

# 3. Criação da Tabela BRONZE no Lakehouse
target_table_name = "dev_catalogue.staging_schema.bronze_insurance_costs"

df_raw_spark.write.format("delta").mode("overwrite").saveAsTable(target_table_name)

print(f"\nSUCESSO: Dados brutos salvos na Tabela Delta: {target_table_name}")
```
